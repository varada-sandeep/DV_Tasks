```python
df_netflix['date_added'] = pd.to_datetime(df_netflix['date_added'], format='%d-%b-%y', errors='coerce')

df_netflix = df_netflix.sort_values('date_added').reset_index(drop=True)

content_over_time = df_netflix['date_added'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(content_over_time.index, content_over_time.values)
plt.title('Content Added Over Time')
plt.xlabel('Date Added')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
content_per_year = df_netflix['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.scatter(content_per_year.index, content_per_year.values)
plt.title('Content Added per Release Year')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python
import numpy as np
import seaborn as sns

movie_durations = df_netflix[df_netflix['type'] == 'Movie']['duration'].str.extract('(\d+)').astype(float).dropna()
plt.figure(figsize=(12, 2))
sns.rugplot(movie_durations, height=0.5)
plt.title('Rug Plot of Movie Durations')
plt.xlabel('Duration (minutes)')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
plt.stem(content_over_time.index, content_over_time.values)
plt.title('Stem Plot of Content Added Over Time (Daily)')
plt.xlabel('Date Added')
plt.ylabel('Number of Titles Added')
plt.show()
```

```python

df_netflix['year_month'] = df_netflix['date_added'].dt.to_period('M')

monthly_counts = df_netflix.groupby(['year_month', 'type']).size().unstack(fill_value=0)

monthly_counts.index = monthly_counts.index.to_timestamp()

plt.figure(figsize=(12, 6))
plt.stackplot(monthly_counts.index, monthly_counts['Movie'], monthly_counts['TV Show'], labels=['Movie', 'TV Show'])
plt.title('Stack Plot of Content Type Added Over Time (Monthly)')
plt.xlabel('Date Added')
plt.ylabel('Number of Titles Added')
plt.legend(loc='upper left')
plt.show()
```
