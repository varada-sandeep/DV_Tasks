```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Sine Wave Plot')
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)
x_scatter = np.random.rand(50) * 10
y_scatter = np.random.rand(50) * 10
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x_scatter, y_scatter, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Scatter Plot')
plt.colorbar(label='Color intensity')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(1, 10, size=len(categories))

plt.bar(categories, values, color='skyblue')
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Simple Bar Chart')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x_seaborn = np.linspace(0, 10, 100)
y_seaborn = np.sin(x_seaborn) + np.random.randn(100) * 0.5

sns.lineplot(x=x_seaborn, y=y_seaborn)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Seaborn Line Plot')
plt.grid(True)
plt.show()
```

```python
import plotly.express as px
import numpy as np

x_plotly = np.random.rand(50)
y_plotly = np.random.rand(50)
size_plotly = np.random.rand(50) * 30

fig = px.scatter(x=x_plotly, y=y_plotly, size=size_plotly, title='Simple Plotly Scatter Plot')
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30, color='lightcoral', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Simple Matplotlib Histogram')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data_seaborn = np.random.normal(loc=0, scale=1, size=100)
data_seaborn = np.append(data_seaborn, np.random.normal(loc=3, scale=1.5, size=50))

sns.boxplot(y=data_seaborn)
plt.ylabel('Value')
plt.title('Simple Seaborn Box Plot')
plt.show()
```
